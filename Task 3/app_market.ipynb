## Access the initial data

```python
df.head()
```

## Handling missing values in rating and size column

```python
# Get average of rating and average of size to fill the missing values
avg_rating = df['rating'].mean().round(1)
avg_size = df['size'].mean().round(2)
```

## Filling the missing values in rating and size column

```python
df['rating'] = df['rating'].fillna(avg_rating)
df['size'] = df['size'].fillna(avg_size)
df[['size', 'rating']].isna().sum()
```

The missing values in rating and size columns are properly handled.

## Handling missing values in current_version and andriod_version

```python
# Get most_frequent values of current_version and andriod_version to fill the missing values
most_frequent_current_version = df['current_version'].mode().values[0]
most_frequent_andriod_version = df['andriod_version'].mode().values[0]
```

## Filling missing values in current_version and andriod_version columns

```python
df['current_version'] = df['current_version'].fillna(most_frequent_current_version)
df['andriod_version'] = df['andriod_version'].fillna(most_frequent_andriod_version)
df[['current_version', 'andriod_version']].isna().sum()
```

The missing values in current_version and andriod_version columns are properly handled.

# 3. Metrics Analysis

The distribution of app ratings reveals that most apps have ratings clustered around 4.0 to 4.6, indicating generally positive user feedback across the dataset.

## Check average rating and reviews among categories

The Social category leads with the highest average rating and reviews, followed by Communication, indicating strong user satisfaction and engagement in these categories.








## Average amount spent on paid apps

```python
paid_apps = df[df['type'] == 'Paid']
paid_avg = paid_apps.groupby('category')['price'].mean()
paid_avg
```

## Visualise the average amount spent on paid apps

```python
plt.subplots(figsize = (20, 8))

plt.xlabel('Types of Categories')
plt.ylabel('Average Price')
plt.title('Average Price of Paid Apps by Category')

plt.xticks(rotation = 55)
plt.bar(paid_avg.index, paid_avg.values)
plt.show()
```

Based on the analysis, categories like Finance (average price: 170.64) and Lifestyle (average price: 124.26) have the highest average prices, while Libraries and Demo (0.99) and Personalization (1.86) have the lowest.








## Check relationship between installs and ratings

```python
plt.subplots(figsize = (8, 6))

plt.xlabel('Number of Installs')
plt.ylabel('Rating')
plt.title('Relationship between Installs and Ratings')

plt.scatter(df['rating'], df['installs'], alpha = 0.5, color = 'orange')
plt.show()
```

The scatter plot shows a weak relationship between the number of installs and app ratings, suggesting that high installs do not necessarily correlate with higher ratings, as apps with a wide range of installs have diverse ratings.m

## Check comparison of installs for Free vs. Paid Apps

```python
installs_type = df.groupby('type')['installs'].sum()
installs_type
```

## Visualize the comparison of installs_type

```python
plt.subplots(figsize = (12, 6))

types = ['Free', 'Paid']
plt.title('Proportion of Total Installs for Free vs. Paid Apps')

plt.pie(installs_type, labels = types, colors = ['lightgreen', 'blue'], autopct = '%1.1f%%')
plt.axis('equal')
plt.legend()
plt.show()
```

The analysis reveals that free apps dominate the market, accounting for approximately 99.9% of total installs, while paid apps represent only about 0.1% of installs.

# 4. Sentiment Analysis

## Load the user_reviews dataset

```python
user_reviews = pd.read_csv('user_reviews.csv')
user_reviews
```

## Renaming the columns in the dataset

```python
cols = {'App': 'app',	
        'Translated_Review': 'translated_review',
        'Sentiment': 'sentiment',
        'Sentiment_Polarity': 'sentiment_polarity',
        'Sentiment_Subjectivity': 'sentiment_subjectivity'}
user_reviews.rename(columns = cols, inplace = True)
```

## Check for missing values count in the dataframe

```python
user_reviews.isna().sum()
```

## Handling missing values by dropna() method

```python
reviews = user_reviews.dropna()
reviews
```

## Check missing values in reviews dataframe

```python
reviews.isna().sum()
```

The missing values in user_reviews dataframe are properly handled.

## Access the initial data

```python
reviews.head()
```

## Check the distribution of sentiment labels

```python
sentiment = reviews['sentiment'].value_counts()
sentiment
```

## Visualise the distribution of sentiment labels

```python
plt.xlabel('Sentiment Label')
plt.ylabel('Distribution Count')
plt.title('Distribution of Sentiment Labels')

plt.bar(sentiment.index, sentiment.values)
plt.show()
```

The analysis reveals that positive reviews dominate the dataset, followed by negative and neutral reviews, suggesting a largely favorable sentiment among users.

## Ignore Warnings

```python
import warnings
warnings.filterwarnings('ignore')
```

## Visualise the sentiment polarity by sentiment

```python
plt.xlabel('Sentiment')
plt.ylabel('Sentiment Polarity')
plt.title('Sentiment Polarity Distribution')

sns.boxplot(x = 'sentiment', y = 'sentiment_polarity', data = reviews, palette = 'Blues')
plt.show()
```

The boxplot analysis reveals that positive reviews tend to have higher sentiment polarity scores, while negative reviews are generally lower, with neutral reviews displaying a wider range of polarity values.

## Merge the two datasets 

```python
merged_df = pd.merge(df, user_reviews, on = 'app', how='inner')
merged_df.head()
```

## Calculate the average sentiment polarity for each category

```python
avg_sentiment = merged_df.groupby('category')['sentiment_polarity'].mean()
avg_sentiment.sort_values(ascending = False)
```

## Visualise the top 10 average sentiment polarity

```python
top_10 = avg_sentiment.head(10) 

plt.xlabel('Average Sentiment Polarity')
plt.ylabel('Category')
plt.title('Average Sentiment Polarity by Category (Top 10)')

plt.barh(top_10.index, top_10.values)
plt.show()
```

The analysis reveals that the Comics category has the highest average sentiment polarity, followed by Events and Auto and Vehicles.
